```python
def sigma68(data,ax): return 0.5*(np.nanquantile(data,q = 0.84,axis = ax) - np.nanquantile(data,q = 0.16,axis = ax))
```

```python
dropz = [2.25,3,4]

fractional_errors=np.zeros((len(dropz), 30,42))
emu_p1derrs=np.zeros((len(dropz), 30,42))
emu_p1ds = np.zeros((len(dropz), 30,42))
true_p1ds=np.zeros((len(dropz), 30,42))



for i,z in enumerate(dropz):
    emulator = emulator_train(emuparams, zmax=4.5, Nsim=30, nepochs=200, step_size = 150, postprocessing='768', kmax_Mpc=4,initial_weights=initial_weights, drop_sim=None, drop_z=z,list_archives=['data_input_axes','data_input_phases'], ndeg=5)

    truth=pd_archive.archivePD(z_max=4.5,nsamples=30)
    truth.average_over_samples(flag="all")
    truth = [d for d in truth.data_av_all if ((d['z'] == z)&(d['ind_tau'] == 0))]
    
    fractional_error, emu_p1d, true_p1d, emu_p1derr = test_sim(emulator.emulator, truth, emuparams, emulator.paramLims, emulator.device, emulator.yscalings,  kmax_Mpc_test=4, Nz=30, ndeg=5 )

    fractional_errors[i] = fractional_error
    emu_p1ds[i] = emu_p1d
    true_p1ds[i] = true_p1d
    emu_p1derrs[i] = emu_p1derr
    
    
```
